```python
import pandas as pd
import pandas_gbq
df = pandas_gbq.read_gbq(query)
```

```python
df.to_csv(r'sstAll.csv',index=False)
```

```python
dfDownload = df
dfDownload
```

```python
dfGroupYearSST=dfDownload.groupby(['year']).agg({'sea_surface_temp': ['max']})
dfGroupYearSST
```

```python
dfGroupYearSST.plot.bar(figsize=(10,5), color = "navy")
plt.legend("")
plt.title("Sea Surface Temperature Maximum by Year (1851-2017)")
plt.xticks([])
plt.savefig("sst_max_by_year")
```
